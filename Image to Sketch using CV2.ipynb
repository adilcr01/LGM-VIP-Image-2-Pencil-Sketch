```python
grey_image = cv2.cvtColor(Original_image,cv2.COLOR_BGR2GRAY)
#plt.figure(figsize=(10,10))
#plt.imshow(grey_image)
#plt.title('Gray Scale Image')
#plt.show()
display(Image.fromarray(grey_image))
```

## Made a function taht take your Image path as Input and give you Pencil Sketch of your Image as Output

```python
def image2sketch(img_path):
    image=cv2.imread(img_path)

    grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

    invert_image=cv2.bitwise_not(grey_image)

    blur_image=cv2.GaussianBlur(invert_image,(21,21),0)#ksize=(21,21) value range from 1 to 21 

    inverted_blur_image=cv2.bitwise_not(blur_image)

    sketch_image=cv2.divide(grey_image,inverted_blur_image,scale=256.0)
    #scale --> if scale value increase, Image become more white

    cv2.imwrite(img_path.split('.')[0]+'_sketch.png',sketch_image)
```

```python
image2sketch('Dog2.jpg')
```
